```python
from filters import compute_h_z, compute_b_a
from plots import show_surface, show_path, show_ir
import numpy as np
```

```python
unit_circle = np.exp(1j*np.linspace(0, 2*np.pi, 1000))
```

```python
# Low pass
zeros = np.array([0.7])
poles = np.array([])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
show_ir(zeros, poles)
```

```python
# High pass
zeros = np.array([0.4, 0.4*np.exp(1j*np.pi/4)])
poles = np.array([0.8])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
show_ir(zeros, poles)
```

```python
# High shelve
zeros = np.array([-0.8])
poles = np.array([-0.9])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
show_ir(zeros, poles)
```

```python
# Stop band
zeros = np.array([0.9*np.exp(1j*np.pi/4)])
poles = np.array([0.8*np.exp(1j*np.pi/4)])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
show_ir(zeros, poles)
```

```python
# Resonance
zeros = np.array([0.8*np.exp(1j*np.pi/4)])
poles = np.array([0.9*np.exp(1j*np.pi/4)])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
show_ir(zeros, poles)
```
