```python
from filters import compute_h_z, compute_b_a
from plots import show_surface, show_path
import numpy as np
import matplotlib.pyplot as plt
```

```python
unit_circle = np.exp(1j*np.linspace(0, 2*np.pi, 1000))
```

```python
# One-Pole Low pass
zeros = np.array([])
poles = np.array([0.9])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```

```python
# High pass, one pole one zero
zeros = np.array([0.9])
poles = np.array([0.8])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```

```python
# Low shelve
zeros = np.array([0.8])
poles = np.array([0.9])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```

```python
# High shelve
zeros = np.array([-0.8])
poles = np.array([-0.9])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```

```python
# Hole
zeros = np.array([0.9*np.exp(1j*np.pi/4)])
poles = np.array([0.8*np.exp(1j*np.pi/4)])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```

```python
# Peak
zeros = np.array([0.8*np.exp(1j*np.pi/4)])
poles = np.array([0.9*np.exp(1j*np.pi/4)])
show_surface((-2, 2), zeros, poles)
show_path(unit_circle, zeros, poles)
```
